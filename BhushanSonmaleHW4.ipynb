```python
class Vehicle:
    
    #Constructor to accept values of all instance variables
    def __init__(self, make, model, yr_mfd, mil):
        self.__veh_make = make
        self.__veh_model = model
        self.__veh_yr_mfd = yr_mfd
        self.__veh_mil = mil
        
    #Defining accessor and mutator methods
    def get_make(self):
        return self.__veh_make
    def get_model(self):
        return self.__veh_model
    def get_yr_mfd(self):
        return self.__veh_yr_mfd
    def get_mil(self):
        return self.__veh_mil
    
    def set_make(self, make):
        self.__veh_make = make
    def set_model(self, make):
        self.__veh_model = model
    def set_yr_mfd(self, make):
        self.__veh_yr_mfd = yr_mfd
    def set_mil(self, make):
        self.__veh_mil = mil
    
    #Redefining __str__() method to return a single string
    def __str__(self):
        return "Make: " + self.__veh_make + "; Model: " + self.__veh_model + "; Year of Manufacture: " + str(self.__veh_yr_mfd) + "; Mileage: " + str(self.__veh_mil)
    
```

```python
from employee import Employee
class FullTimeEmployee(Employee):
    
    def __init__(self, ename, eadd, veh, sal):
        super().__init__(ename, eadd, veh)
        self.__salary = sal
        
    def get_salary(self):
        return self.__salary
    
    def set_salary(self, sal):
        self.__salary = sal
        
    def cal_compensation(self):
        if(self.__salary <= 45000):
            compensation = self.__salary - (0.18 * self.__salary)
            return compensation
        else:
            if(self.__salary > 45000 and self.__salary <= 82000):
                tx1 = 0.18 * 45000
                tx2 = 0.28 * (self.__salary - 45000)
                compensation = self.__salary - (tx1 + tx2)
                return compensation
            else:
                tx1 = 0.18 * 45000
                tx2 = 0.28 * (82000 - 45000)
                tx3 = 0.33 * (self.__salary - 82000)
                compensation = self.__salary - (tx1 + tx2 + tx3)
                return compensation
            
    def __str__(self):
        EmpDataStore = super().__str__()
        EmpDataStore = "\nDetails of Full Time Employee are: " + EmpDataStore
        EmpDataStore += "\nSalary: " + '{0}'.format(str("{0:.2f}".format(self.__salary)))
        return EmpDataStore
```

```python
from employee import Employee
class HourlyEmployee(Employee):
    
    def __init__(self, ename, eadd, veh, hoursWorked, hourlyRate):
        super().__init__(ename, eadd, veh)
        self.__hWorked = hoursWorked
        self.__hRate = hourlyRate
        
    def get_hWorked(self):
        return self.__hWorked
    def get_hRate(self):
        return self.__hRate
    
    def set_hWorked(self, hoursWorked):
        self.__hWorked = hoursWorked
    def set_hRate(self, hourlyRate):
        self.__hRate = hourlyRate
        
    def cal_compensation(self):
        if(self.__hWorked <= 40):
            compensation = self.__hWorked * self.__hRate
            return compensation
        else:
            comp1 = 40 * self.__hRate
            comp2 = (self.__hWorked - 40) * self.__hRate * 1.8
            compensation = comp1 + comp2
            return compensation
        
    def __str__(self):
        EmpDataStore = super().__str__()
        EmpDataStore = "\nDetails of Hourly Employee are: " + EmpDataStore
        EmpDataStore += "\nHours Worked: " + str(self.__hWorked) + "\nHourly Rate: " + str(int(self.__hRate))
        return EmpDataStore
```

```python
from employee import Employee
class Consultant(Employee):
    
    def __init__(self, ename, eadd, veh, hoursWorked, ProjectType):
        super().__init__(ename, eadd, veh)
        self.__hWorked = hoursWorked
        self.__pType = ProjectType
        
    def get_hWorked(self):
        return self.__hWorked
    def get_pType(self):
        return self.__pType
    
    def set_hWorked(self, hoursWorked):
        self.__hWorked = hoursWorked
    def set_pType(self, ProjectType):
        self.__pType = ProjectType
        
    
    def cal_compensation(self):
        if(self.__pType == 1):
            compensation = self.__hWorked * 55.0
            return compensation
        if(self.__pType == 2):
            compensation = self.__hWorked * 70.0
            return compensation
        if(self.__pType == 3):
            compensation = self.__hWorked * 85.0
            return compensation
        
    def __str__(self):
        EmpDataStore = super().__str__()
        EmpDataStore = "\nDetails of Consultant are: " + EmpDataStore
        EmpDataStore += "\nHours Worked: " + str(self.__hWorked) + "\nProject Type: " + str(self.__pType)
        return EmpDataStore
```

```python
from employee import Employee
from ftemployee import FullTimeEmployee
from hemployee import HourlyEmployee
from consultant import Consultant
from vehicle import Vehicle
import pickle
import sys

def main():
    run_menu_options()
    
def run_menu_options():
    employee_list = read_file_data()
    sel = 1
    
    while sel >= 1 and sel <= 5:
        print("\n==== Menu ====")
        print("1. To add an employee")
        print("2. To print the name and address of all employees")
        print("3. To print the employee name and compensation of all employees")
        print("4. To print the employee name with vehicle information for high mileage vehicles")
        print("5. To exit program")
        
        try:
            sel = int(input("Your selection: "))
            if sel < 1 or sel > 5:
                raise ValueError
        except ValueError:
            print('You must enter an integer between 1 and 5!')
            print('Try again')
            sel = 1
            continue
            
        if sel == 1:
            employee_list.append(run_option1())
        elif sel == 2:
            if len(employee_list) == 0:
                print("No employees in database!")
            else:
                run_option2(employee_list)
        elif sel == 3:
            if len(employee_list) == 0:
                print("No employees in database!")
            else:
                run_option3(employee_list)
        elif sel == 4:
            if len(employee_list) == 0:
                print("No employees in database!")
            else:
                run_option4(employee_list)
        elif sel == 5:
                run_option5(employee_list)
            

def read_file_data():
    my_lst = [] #Create an empty list
    try:
        #Call the open command within a try block.  Note that the file is opend for binary read.
        fb = open('employeedata.dat', 'rb') 
    except FileNotFoundError:
        #If the file is not there (this will be the case, the first time the program is executed), the empty file is returned.
        return my_lst 
    #If there is no error reading the file, the rest of the code is executed.
    else: 
        try:
            #This is a repeated loop which will terminate only when the EOFError occurs.
            while (True): 
                #This statement reads the next object from the file and appends it to the list
                my_lst.append(pickle.load(fb))
        #When the EOFError occurs, the file is closed and the list (containing all the book objects that have just been read) is returned.
        except EOFError: 
            fb.close()
            return my_lst
        

def run_option1():
    sel = 0
    while sel < 1 or sel > 3: #Repeat the prompt until the user enters an valid value of 1, 2 and 3
        try:
            sel = int(input("Type of Employee?(1-Full Time;2-Hourly;3-Consultant) : "))
            if(sel < 1 or sel > 3):
                raise ValueError
        except ValueError: 
            print('Wrong selection!')
            continue
    
    ename, eadd = get_employee_input()
    make, model, yr_mfd, mil = get_vehicle_input()
    a_veh = Vehicle(make, model, yr_mfd, mil)
    
    if sel == 1:
        sal = get_ftemp_input()
        an_emp = FullTimeEmployee(ename, eadd, a_veh, sal)
    elif sel ==2:
        hoursWorked, hourlyRate = get_hemp_input() 
        an_emp = HourlyEmployee(ename, eadd, a_veh, hoursWorked, hourlyRate)
    elif sel ==3:
        hoursWorked, projectType = get_consultant_input() 
        an_emp = Consultant(ename, eadd, a_veh, hoursWorked, projectType)
    return an_emp

def run_option2(my_lst):
    for obj in my_lst:
        print(obj.__str__())

def run_option3(my_lst):
    print('\nEmployee name and Compensation of all Employees')
    print("================================")
    for obj in my_lst:
        comp = obj.cal_compensation()
        print('{0}\'s is ${1:.2f}'.format(obj.get_emp_name(), comp))
        
def run_option4(my_lst):
    print('\nEmployee name and Vehicle Data For High Mileage Vehicles')
    print("================================")
    for obj in my_lst:
        obj.get_option4(my_lst)

def run_option5(my_lst): 
    print("You chose to exit the program")
    yes_no = input("Are you sure (Y/N)? ")

    if yes_no.lower() == 'y':
        fb = open('employeedata.dat', 'wb')
        for obj in my_lst:
            pickle.dump(obj, fb)
        fb.close()
        sys.exit()

def get_employee_input():
    ename = input('Enter the employee\'s name: ')
    eadd = input('Enter the employee\'s address: ')
    return (ename, eadd)

    
def get_vehicle_input():
    make = input('Enter the vehicle make: ')
    model = input('Enter the vehicle model: ')
    err_flg = True	
    while err_flg:
        try:
            yr_mfd = int(input('Enter the year of manufacture: '))
        except ValueError:
            print('\nEnter an integer value for year!!')
        else:
            err_flg = False
    err_flg = True
    while err_flg:
        try:
            mil = int(input('Enter the mileage: '))
        except ValueError:
            print('\nEnter an integer value for mileage!!')
        else:
            err_flg = False
    return (make, model, yr_mfd, mil)

    
def get_ftemp_input():
    err_flg = True
    while err_flg:
        try:
            sal = float(input('Enter the salary: '))
        except ValueError:
            print('\nEnter a numerical value for the salary!!')
        else:
            err_flg = False
    return (sal)

def get_hemp_input():
    err_flg = True
    while err_flg:
        try:
            hoursWorked = int(input('Enter the hours worked: '))
        except ValueError:
            print('\nEnter a numerical value for the hours worked!!')
        else:
            err_flg = False
    err_flg = True
    while err_flg:
        try:
            hourlyRate = float(input('Enter the hourly rate: '))
        except ValueError:
            print('\nEnter a numerical value for the hourly rate!!')
        else:
            err_flg = False
    
    return(hoursWorked, hourlyRate)

def get_consultant_input():
    err_flg = True
    while err_flg:
        try:
            hoursWorked = int(input('Enter the hours worked: '))
        except ValueError:
            print('\nEnter a numerical value for the hours worked!!')
        else:
            err_flg = False
    err_flg = True
    
    while err_flg:
        try:
            projectType = int(input('Project Type? (Enter a number between 1 and 3): '))
            if(projectType < 1 or projectType > 3):
                raise ValueError
        except ValueError:
            print('Enter a value of between 1 and 3 only!!')
        else:
            err_flg = False
    err_flg = True   
    return (hoursWorked, projectType)


main()
```
