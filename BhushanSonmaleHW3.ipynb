```python
'''
1. Write a Python program to add 'ing' at the end of a given string (length
should be at least 3). If the given string already ends with 'ing' then add 'ly'
instead. If the string length of the given string is less than 3, leave it
unchanged.
'''

my_str = input('Enter a string: ')
str_len = len(my_str)
if str_len<3:
    print(my_str)
else:
    if my_str.endswith('ing'):
        print(my_str+'ly')
    else:
        print(my_str+'ing')
```

```python
'''
2. Write a Python function called has_sublist (), that accepts a list as input
and checks if any of its elements are themselves lists.
'''

def has_sublist(input_list):
    for item in input_list:
        if type(item) == list:
            val = 'true'
            break;
        else:
            val = 'false'
    print('It is', val, 'that the list',input_list,'has a sublist')

def main():
    test = [3,8,5,4,[5,7]]
    has_sublist(test)
    
main()
```

```python
'''
3. Write a Python function concat_list() that accepts a list containing string
values and an integer as input and prints out a new list where each
element in the list is concatenated with all integers from 1 to n.
'''

def concat_list(input_list, n):
    out_list = []
    for item in input_list:
        for ele in range(1,n+1):
            out_list.append(item+str(ele))
    print(out_list)

def main():
    input_list = ['a','x','d']
    n = 5
    concat_list(input_list,n)
    
main()
```

```python
'''
4. Write a Python program that accepts a comma separated sequence of
words as input and prints the unique words in sorted form
(alphanumerically)
'''

seq = input('Enter a sequence of words separated by commas:')
old = seq.split(',')
new = []
for item in old:
    if item not in new:
        new.append(item)
new.sort()
print(new)
```

```python
'''
5. Write a Python program to create a dictionary from a string by tracking the
count of characters in a string. Both upper case and lower-case alphabets
should be treated as one. Do not count anything but letters of the alphabet.
That is, do not count spaces, or any other special characters.
'''
import re
my_str = input('Enter a string: ')
for k in my_str.split("\n"):
    str2 = "".join(re.findall(r"[a-zA-Z]+", k))
    upstr = str2.upper()

dict_str = {}
for ele in upstr:
    dict_str.setdefault(ele,0)
    dict_str[ele] += 1
print(dict_str)
```
